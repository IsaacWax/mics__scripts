```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0, 2e-3, 1000)  
Vs = 10 * np.sin(2 * np.pi * 1000 * t)  # 6280

plt.plot(t * 1000, Vs)
plt.title("Source Voltage vs Time")
plt.xlabel("Time (ms)")
plt.ylabel("Voltage (V)")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0, 6e-3, 3000)
Vs = 10 * np.sin(2 * np.pi * 1000 * t)

clip_level = 7.4
Vo = np.clip(Vs, -clip_level, clip_level)

plt.plot(t * 1000, Vs, label="Source Voltage (10 sin 6280t)")
plt.plot(t * 1000, Vo, label="Clipped Output (±7.4 V)", linewidth=2)
plt.title("Zener Clipping at ±7.4 V")
plt.xlabel("Time (ms)")
plt.ylabel("Voltage (V)")
plt.legend()
plt.grid(True)
plt.show()
```
